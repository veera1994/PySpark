```
from pyspark.sql.functions import format_number, min, max, mean, count
```

```
result = df.describe()
```

```
# Changing the datatypes
result.select(result['summary'],
             format_number(result['Open'].cast('float'),2).alias('Open'),
             format_number(result['High'].cast('float'),2).alias('High'),
             format_number(result['Low'].cast('float'),2).alias('Low'),
             format_number(result['Close'].cast('float'),2).alias('Close'),
             result['Volume'].cast('int').alias('Volume'),
             format_number(result['Adj Close'].cast('float'),2).alias('Adj Close')).show()
```

```
# Create a new column called HV Ratio
newdf = df.withColumn('HV Ratio', df['High']/df['Volume'])
newdf.select(newdf['HV Ratio']).show()
```

```
# Peak high in price
df.select(max(df['High'])).show()
```

```
# Mean of close column
df.select(mean(df['Close'])).show()
```

```
df.select(max(df['Volume']), min(df['Volume'])).show()
```

```
# How many days the close lower than 60dollars
df.filter(df['Close']<60).count()
```

```
x1 = df.filter(df['High']>80).count()
x2 = df.count()
x1/x2
```

```
# Pearson correlation
from pyspark.sql.functions import corr
df.select(corr('High', 'Volume')).show()
```

```
from pyspark.sql.functions import year, month
```

```
# Find the max of high for each year
newdf = df.withColumn('Year', year('Date'))
newdf.show()
```

```
newdf.groupBy('Year').max().select(['Year','max(High)']).show()
```

```
# What is the average close of each month
newdf = df.withColumn('Month', month('Date'))
newdf.show()
```

```
newdf.groupBy('Month').avg().select(['Month','avg(Close)']).orderBy('Month').show()
```
